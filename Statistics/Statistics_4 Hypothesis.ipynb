# 1) Z-test
### 1.1) One tail test
#### 1.1.1) Left Tail test
Example - Car mileage test.
- Ho : All our cars average mileage is greater than 25KMPL
- Ha : Average mileage is less than 25kmpl <br>

Left-tailed test: The alternative hypothesis contains the "<" sign

#### 1.1.2)  Right Tail test
Example - New method will reduce delivery time
   - Ho : Our new home delivery method reduces delivery time.
   - Ha : New method will increase the delivery time <br>
   
Right-tailed test: The alternative hypothesis contains the ">" sign

### 1.2) Two tail test
Example - Difference in car mileage when drove by two different persons<br>
- Ho: The two Groups have equal means 
- Ha: The two Groups have different means or not equals

Two-tailed test: The alternative hypothesis contains the "≠" sign

<font face = "Verdana" size = "3"><u>  Real time practicle example on Left Tail test</u>


<font face = "Verdana" size = "3"> ➤ A car manufacturing company claims mileage of  all their cars is 25 kmpl<br>
    <b><mark>(mean / avg of all cars = 25)</b></mark>. <br>
➤<font face = "Verdana" size = "3">The population standard deviation is known from the past experience to be 1.5 kmpl <br><mark> <b> std= 1.5</b>.</mark>
<br>➤A car rental company wants to test the null hypothesis that the average mileage is at least 25 kmpl. Car rental representative drove<mark><b> 30 cars (samples) </b></mark>of the
same company and recorded each car mileage. <br>The <mark><b> average</b></mark> mileage of the sample turned out to be <mark><b>24.2 kmpl.

<font face = "Verdana" size = "3" > <b><i>Hypothesis construction for Left tail test are stated below   </i></b><br>
<font face = "Verdana" size = "3" ><p style="color: red;">(H0): Null Hypothesis >= 25kmpl <br>
(H1): Alternative Hypothesis < 25kmpl

```python
import numpy as np
Population_mean = 25
observed_value= sample_mean  = 24.2
SD_of_Population = 1.5
n = 30
alpha= 0.05
```

```python
Z = (observed_value- Population_mean)/ (SD_of_Population/np.sqrt(30))
Z
```

```python
# Below Z score need to check in the Z-table to get the p-value
```

```python
# This is How we did manually 
```

```python
#Using statistics tools
import scipy.stats as st
import numpy as np
p_value = st.norm.cdf(sample_mean, loc=Population_mean, scale=SD_of_Population/np.sqrt(30))
p_value
```

# Important points
<font face = "Verdana" size = "3"> • Standard alpha level = 0.05  ....|you are allow to make 5% error|<br>
<font face = "Verdana" size = "3">• If p-value is < 0.05 then we reject null hypothesis (Ho)<br>
<font face = "Verdana" size = "3">• If p-value is > 0.05 then we failed to reject the Ho<br>

```python
if p_value < alpha:
    print("Ho is rejected")
else :
    print("Ho is accepted")
```

<blockquote data-sourcepos="3:1-3:258"><font face = "Verdana" size = "3"> Conclusion: as the p_value is less than alpha (0.05), Hence we reject the null hypothesis. Which also means that, all the cars do not give 25 mileage so rental company will not buy the cars.

<font face = "Verdana" size = "4"><b>Type 1 Error | False Positive : </b><br>
<font face = "Verdana" size = "3">In our example, If average car mileage is really 25 or > 25 kmpl and rental agency rejects for buying cars  it. It's Type 1 error.

<font face = "Verdana" size = "4"><b> Type 2 Error | False Negative :</b><br>
<font face = "Verdana" size = "3">If Average car mileage is really < 25 kmpl but rental agency buying cars (failed to reject HO). It's Type 2 error.

<font face = "Verdana" size = "3"> 2.2.1) Independent sample

<font face = "Verdana" size = "3"><u> Real time practicle example on Independent sample ttest</u>

<font face = "Verdana" size = "3">  <p data-sourcepos="3:3-3:325">The teacher was quite upset when she realized her mistake. She had conducted a hypothesis test on the average marks of all students, but she had not considered the gender of the students. She knew that there could be a difference in the average marks of boys and girls, so she decided to perform an <b>independent sample test.</p>
    
<font face = "Verdana" size = "3" > <b><i>Hypothesis construction for independent sample ttest are stated below   </i></b><br>
<font face = "Verdana" size = "3" ><p style="color: red;">(H0): The null hypothesis is that the average marks is the same for both, as stated by the professor<br>(H1): alternative hypothesis is : that the average marks is different for both or ≠ , as stated by the teacher </p>

```python
hyp.head()
```

```python
hyp['Gender'].value_counts()
```

```python
# How to subset the data on Gender
hyp[hyp['Gender']==1].Math1.head()
```

```python
# Independent sample
t,p=st.ttest_ind (hyp[hyp['Gender']==1].Math1, hyp [hyp['Gender']==0]. Math1)
```

```python
p
```

```python
if p < alpha:
    print("Ho is rejected")
else :
    print("Ho is accepted")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import warnings; warnings.filterwarnings("ignore")
# Create a figure with subplots
fig, (ax_box, ax_dist) = plt.subplots(2, 1, figsize=(7, 8))

# Horizontal box plot with PDF
sns.boxplot(y='Gender', x='Math1', data=hyp, orient='h', ax=ax_box)
ax_box.set_xlabel('Math1')
ax_box.set_ylabel('Gender')
ax_box.set_title('Horizontal Box Plot with Probability Density Function')

# Density plots by Gender
sns.distplot(hyp[hyp['Gender'] == 1]['Math1'], hist=False, rug=False, label='Gender 1', ax=ax_dist)
sns.distplot(hyp[hyp['Gender'] == 0]['Math1'], hist=False, rug=False, label='Gender 0', ax=ax_dist)
ax_dist.set_xlabel('Math1')
ax_dist.set_ylabel('Density')
ax_dist.set_title('Density Plot by Gender')
ax_dist.legend()

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

<font face = "Verdana" size = "3"> 2.2.2) paired sample

<font face = "Verdana" size = "3"><u> Real time practicle example on paired sample ttest</u>

<font face = "Verdana" size = "3"> <p data-sourcepos="5:1-5:384">The teacher acknowledged his mistake and requested a second chance to prove himself. He implemented a new teaching technique and conducted a second test one month later. The teacher was confident that his new technique had improved the students' performance, but the professor was not convinced. The professor decided to conduct another hypothesis test to verify the teacher's claims.</p>

<font face = "Verdana" size = "3" > <b><i>Hypothesis construction for paired sample ttest are stated below   </i></b><br>
<font face = "Verdana" size = "3" ><p style="color: red;">(H0) : μ1=μ2.as stated by the professor<br>(H1) : μ1≠μ2 as stated by the teacher </p>

```python
# Paired sample
t,p=st.ttest_rel (hyp.Math1, hyp.Math2)
```

```python
p
```

```python
if p < alpha:
    print("Ho is rejected")
else :
    print("Ho is accepted")
```

```python
# Create a figure with subplots
fig, (ax_box, ax_dist) = plt.subplots(2, 1, figsize=(10, 8))

# Box plots for Math1 and Math2
sns.boxplot(data=hyp[['Math1', 'Math2']], ax=ax_box,orient='h')
ax_box.set_xticklabels(['Math1', 'Math2'])
ax_box.set_ylabel('Scores')
ax_box.set_title('Box Plots for Math1 and Math2')

# Density plots for Math1 and Math2
sns.kdeplot(data=hyp, x='Math1', label='Math1', ax=ax_dist)
sns.kdeplot(data=hyp, x='Math2', label='Math2', ax=ax_dist)
ax_dist.set_xlabel('Scores')
ax_dist.set_ylabel('Density')
ax_dist.set_title('Density Plots for Math1 and Math2')
ax_dist.legend()

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

There is an another test which is more than 2 sample t test. But for each t test we allow 5% error (alpha), so if we conduct more than 2 sample t test we are
adding 5% more error for each extra sample t test. In order to resolve this problem we go for ANOVA test. i.e. Analysis Of variance

The first t-test was a one-sample t-test, which compared the average marks of the students to a hypothesized value of 60. The second t-test was an independent sample t-test, which compared the average marks of boys and girls. The third t-test was a paired sample t-test, which compared the average marks of the students before and after the new teaching technique was implemented.</p>
    
<font face = "Verdana" size = "3"><p data-sourcepos="5:1-5:309">The results of all three t-tests showed that the subject teacher's assumption was wrong. The students did not score an average of 60 on the test, there was a significant difference in the average marks of boys and girls, and the new teaching technique did not have a significant impact on the students' marks.</p>
